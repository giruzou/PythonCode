```python
from IPython.core.pylabtools import figsize
import numpy as np
from matplotlib import pyplot as plt 
%matplotlib inline
figsize(12,4)

```

```python
colors=["#348ABD", "#A60628"]
prior = [1/21., 20/21.]
posterior = [0.087, 1-0.087]
plt.bar([0,0.7], prior, alpha=0.70, width=0.25,
             color=colors[0], label="prior distribution",
              lw="2", edgecolor=colors[0])
plt.bar([0+0.25, 0.7 + 0.25], posterior, alpha=0.7,
               width=0.25, color=colors[1],
               label="posterior distribution",
               lw="2", edgecolor=colors[1])
plt.xticks([0.20, 0.95], ['Librarial', 'Farmer'])
plt.legend(loc="upper left")
plt.ylabel("Probability")
plt.title("Prior and Posterior probablities of Steve's occupation")
```

```python
import scipy.stats as stats
figsize(12,4)

poisson=stats.poisson
lambda_=[1.5,4,25]
colors=["#348ABD", "#A60628"]

a= np.arange(16)
plt.bar(a,poisson.pmf(a,lambda_[0]), color=colors[0],
                label=r"$\lambda = %.1f$" % lambda_[0],
                alpha=0.6, edgecolor=colors[0],lw="3")
plt.bar(a,poisson.pmf(a,lambda_[1]), color=colors[1],
                label=r"$\lambda = %.1f$" % lambda_[1],
                alpha=0.6, edgecolor=colors[1],lw="3")
plt.xticks(a,a)
plt.legend()
plt.ylabel(r"Probablity of $k$")
plt.xlabel(r"$k$")
plt.title("Probability mass function of a Poisson random variables,"
              r"differing $\lambda$ values")
```

```python
a=np.linspace(0, 4, 100)
expo=stats.expon
lambda_ = [0.5, 1]
colors=["#348ABD", "#A60628"]
for l, c in zip (lambda_, colors):
    plt.plot(a, expo.pdf(a, scale=1. / l),
                  lw=3, color=c, label=r"$\lambda = %.1f$" % 1)
    plt.fill_between(a,expo.pdf(a, scale=1. / l),
                    color=c, alpha=.33)

plt.legend()
plt.ylabel("Probablity density function at $z$")
plt.xlabel("$z$")
plt.ylim(0, 1.2)
plt.title("Probablity density function of an exponential "
             "random variable, differing $\lambda$ values")
```
