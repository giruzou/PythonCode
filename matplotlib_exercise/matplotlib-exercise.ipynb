```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
xs=np.linspace(0,2*np.pi,100)
ys=np.sin(xs)
```

```python
#subplots は一回の関数の呼び出しでFigureとAxesインスタンスを生成する．
```

```python
fig,ax=plt.subplots()
ax.plot(xs,ys)
ax.plot(xs,2*ys)
```

# グラフの種類

matplotlib には Axesインスタンスメソッドとして，さまざまな種類のグラフの作成機能が実装されている．

```python
N=15
xs=np.linspace(0,2*np.pi,N)
ys=np.sin(xs)
ys_err=0.5*np.random.rand(N)
fig,ax=plt.subplots(2,2,figsize=(8,4),sharey=True)
ax[0][0].step(xs,ys)
ax[0][1].errorbar(xs,ys,yerr=ys_err,fmt='o')
ax[1][0].fill(xs,ys)
ax[1][1].bar(xs,ys,0.2)
```

```python
import matplotlib 
matplotlib.style.use('ggplot')
N=50
xs=np.random.rand(N)
ys=np.random.rand(N)
z=np.sqrt(xs**2+ys**2)

fig,ax=plt.subplots()
ax.scatter(xs,ys,s=80,c=z)
```

# グラフの一部を拡大

```python
xs=np.linspace(0,2*np.pi,100)
ys=np.sin(xs)

fig,ax=plt.subplots(subplot_kw=dict(axisbg='#eeeeff'))
ax.plot(xs,ys)

axins=fig.add_axes([0.55,0.55,0.3,0.3],axisbg='none')
axins.plot(-xs,ys)
```

from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

```python
n=200
xs=np.linspace(0,2*np.pi,n)
ys=np.cos(2*np.pi*xs) * np.exp(-xs) + 0.1 * np.random.rand(n)
fig,ax=plt.subplots(subplot_kw=dict(axisbg='#eeeeff'))
ax.plot(xs,ys)

#拡大図を生成
axins=zoomed_inset_axes(ax,4,loc=9,axes_kwargs=dict(axisbg='none'))
axins.plot(xs,ys)
x1,x2,y1,y2=0.7,1.3,0.3,0.5
axins.set_xlim(x1,x2)
axins.set_ylim(y1,y2)
#拡大図のメモリのラベルを非表示
axins.set_xticklabels([])
axins.set_yticklabels([])
mark_inset(ax,axins,loc1=2,loc2=4,fc='none',ec='0.5')
```
