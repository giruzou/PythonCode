```python
import time
from IPython.display import  clear_output
from ipythonblocks import BlockGrid,colors
```

```python
grid = BlockGrid(width=5,height=5,fill=colors['Purple'])
grid.show()
```

```python
grid[0,0]=colors['Lime']
grid[-1,0]=colors["Lime"]
grid[:,-1]=colors['Lime']
grid.show()
```

```python
n=5
grid=BlockGrid(width=2*n+1,height=2*n+1,fill=colors["White"])
A=grid[n+1:,:n]
B=grid[:n,n+1:]
C=grid[n+1:,n+1:]
A[:,:]=colors["Cyan"]
B[:,:]=colors["Lime"]
C[:,:]=colors["Yellow"]
grid.show()
```

```python
for i in range(n):
    for j in range(n):
        #we reset the matrix clolors.
        A[:,:]=colors["Cyan"]
        B[:,:]=colors["Lime"]
        C[:,:]=colors["Yellow"]
        #we highlight the adeauate rows
        #and colomns in red.
        A[i,:]=colors['Red']
        B[:,j]=colors['Red']
        C[i,j]=colors['Red']
        #we animate the grid in the loop.
        #clear_output()
        grid.show()
        time.sleep(1)
```

```python
from PIL import Image
imdata=Image.open('screenshot.png').getdata()
row,cols=imdata.size
print(row)
grid=BlockGrid(width=row,height=cols,block_size=4,lines_on=False)
for block,rgb in zip(grid,imdata):
    grid.rgb=rgb[:3]
grid.show()
print("finish")
```
