```python
from scipy import misc
from scipy.ndimage import gaussian_filter,uniform_filter
from matplotlib import pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets
%matplotlib inline
```

```python
def showimg(img):
    plt.imshow(img,cmap=plt.cm.gray,vmin=0,vmax=255)
    plt.axis('off')
```

```python
sample=misc.face(gray=True)
showimg(sample)
```

```python
sample=misc.face(gray=True)
blurred=gaussian_filter(sample,sigma=3)
showimg(blurred)
```

```python
sample=misc.face(gray=True)
uniform=uniform_filter(sample,size=11)
showimg(uniform)
```

```python
def sharptest(alpha,sigma):
    orig=misc.face(gray=True).astype(float)
    blurred=gaussian_filter(orig,sigma=sigma)
    filtered=gaussian_filter(blurred,sigma=1)
    sharped=blurred+alpha*(blurred-filtered)
    fig=plt.figure(figsize=(10,5))

    ax1=fig.add_subplot(131)
    ax2=fig.add_subplot(132)
    ax3=fig.add_subplot(133)
    ax1.axis('off');ax2.axis('off');ax3.axis('off')
    ax1.set_title('orig')
    ax2.set_title('sigma {}'.format(sigma))
    ax3.set_title('alpha {}'.format(alpha))
    ax1.imshow(orig,cmap='gray',vmin=0,vmax=255)
    ax2.imshow(blurred,cmap='gray',vmin=0,vmax=255)
    ax3.imshow(sharped,cmap='gray',vmin=0,vmax=255)
    plt.show()
```

```python
alpha_slider=widgets.IntSlider(value=30,min=20,max=50)
sigma_slider=widgets.IntSlider(value=5,min=1,max=20)
interact(sharptest,alpha=alpha_slider,sigma=sigma_slider)
```
