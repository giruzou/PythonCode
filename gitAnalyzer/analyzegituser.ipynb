# analyze git user

```python
from pprint import pprint
"""
If one use this script with python3,
you must install 'requests' package in advance.
$ pip install requests
"""
import requests
```

```python
user='terasakisatoshi'
response=requests.get('https://api.github.com/users/%s/repos' % user)

pprint(response.json())
```

```python
for idx,repo in enumerate(response.json()):
    if repo['language']=='Python':
        print(idx,repo['name'])
    
```

```python
for repo in response.json():
    if 'PythonCode' in repo['name']:
        print(repo['language'])
```

```python
for repo in response.json():
    print(repo['name'],"=",repo['language'])
```
