# 数値最適化

```python
import numpy as np
import scipy as sp
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f= lambda x : np.cos(x)-x
```

```python
x=np.linspace(-5,5,1000)
y=f(x)
plt.plot(x,y)
plt.axhline(0,color='k')
plt.xlim(-5,5)
```

```python
plt.plot(x,y)
plt.axhline(0,color='k')
plt.scatter([opt.bisect(f,-5,5)],[0],c='r' , s=100)
plt.xlim(-5,5)
```

# minimize

```python
f=lambda x : 1-np.sin(x)/x
x=np.linspace(-20,20,1000)
y=f(x)
plt.plot(x,y)
```

```python
x0=3
xmin=opt.minimize(f,x0).x
plt.plot(x,y)
plt.scatter(x0,f(x0),marker='o',s=300)
plt.scatter(xmin,f(xmin),marker='v',s=300)
plt.xlim(-20,20)
```

```python
x0=10
xmin=opt.minimize(f,x0).x
plt.plot(x,y)
plt.scatter(x0,f(x0),marker='o',s=300)
plt.scatter(xmin,f(xmin),marker='v',s=300)
plt.xlim(-20,20)
```

```python
xmin=opt.minimize(f,x0,method='anneal').x
plt.plot(x,y)
plt.scatter(x0,f(x0),marker='o',s=300)
plt.scatter(xmin,f(xmin),marker='v',s=300)
plt.xlim(-20,20)
```

```python
def g(X):
    x,y=X
    return (np.sin(3*np.pi*x)**2+
                (x-1)**2*(1+np.sin(3*np.pi*y)**2)+
                (y-1)**2*(1+np.sin(2*np.pi*y)**2))
```

```python
n=200
k=10
X,Y=np.mgrid[-k:k:n*1j,-k,k:n*1j]
Z=g(np.vstack((X.ravel(),Y.ravel()))).reshape(n,n)
plt.imshow(np.log(X),cmap=plt.cm.hot_r)
plt.xtick([]);plt.yticks([])
```
